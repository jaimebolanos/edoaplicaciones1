<p style="text-align: justify;"> <span style="color:green">Problema 5</span></p>
 
La expresión que modela la carga y la corriente pra un circuito para un circuito $RC$ son respectivamente:

$$q(t)=\left ( \frac{5}{26}sen(t)-\frac{25}{26 }cos(t)\right )+\frac{285}{26}~e^{-\frac{1}{5}~t}$$
 y 
 
$$i(t)=\frac{5}{26}~cos(t)+\frac{25}{26}sen(t)+\frac{57}{26} ~e^{-\frac{1}{5}~t}$$
Construya las gráficas de $i(t)$ y $q(t)$ para valores de  $0\leqslant t\leqslant 100$ 

<p style="text-align: justify;"> <span style="color:blue">Solución:</span></p>



```python
from sympy import symbols, Function
from sympy import Rational
from sympy import *
from sympy import init_printing 
init_printing()
t,x,y=symbols('t x y')
q=Function('q')
i=Function('i')
```

```python
#Generamos las dos ecuaciones patra graficar

Q=Rational(5,26)*sin(t)-Rational(25,26)*cos(t) +Rational(285,26)*(exp((Rational(-1,5))*t))
R=Eq(q(t),Q)
S=Rational(5,26)*cos(t)+Rational(25,26)*sin(t)-Rational(57,26) *(exp(Rational(-1,5)*t))
T=Eq(i(t),S)
T
```

Enseguida haremos las gráficas de dos maneras diferentes. Una de ellas implica el manejo del M+odulo de Sympy y la otra del módulo Matplotlib. empezaremos con el primero.

```python
from sympy.plotting import plot 
p1=plot(Q,(t,0,100), ylabel='carga $q(t)$', xlabel='tiempo $t$',line_color='red' )
p2=plot(S,(t,0,100), ylabel='corriente $i(t)$', xlabel='tiempo $t$',line_color='blue' )

```

Ahora de la segunda manera

```python
import numpy as np
from numpy import e, pi, sin, exp, cos
import matplotlib.pyplot as plt
def Q(t):
    return 5/26*sin(t)-25/26 * cos(t)+285/26*exp(-t/5)
def i(t):
    return 25/26*sin(t)+5/26 * cos(t)-57/26*exp(-t/5)
def g(t):
    return sin(t) * cos(1/(t+0.1))
def g(t):
    return sin(t) * cos(1/(t))
fig = plt.figure(figsize=(12, 8))
t = np.arange(0, 100, .1)
sub1 = fig.add_subplot(221,ylabel='carga $q(t)$', xlabel='tiempo $t$') # en lugar de plt.subplot(2, 2, 1)
sub1.set_title('CARGA SOBRE EL CONDENSADOR') # non OOP: plt.title('The function f')
sub1.plot(t, Q(t))
plt.grid(True) 
plt.tight_layout()
sub2 = fig.add_subplot(222, xlabel='tiempo $t$',ylabel='corriente $i(t)$')
sub2.set_title('CORRIENTE EN EL CIRCUITO')
sub2.plot(t, i(t))
plt.grid(True) 
plt.show()
```
