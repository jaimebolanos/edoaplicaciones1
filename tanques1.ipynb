<p style="text-align: justify;"> <span style="color:green">Problema 1. Caso: $\dot{v}_1>\dot{v}_2$</span></p>
 
 Un tanque con capacidad de  $50$ litros ($v_T$) contiene inicialmente 500 $g$ de sal en un volumen de 25 litros de solución  (${v_0}$). La solución se mantiene bien mezclada y comienza a entrar una corriente con una concentración de $c_1=5 ~~ g$ a una  velocidad de $\dot{v}_1=9 ~~ \frac{lt}{min}$ y la mezcla sale del tanque a  $\dot{v}_2=7 ~~ \frac{lt}{min}$. Ver la siguiente figura.

![hhhhh](images\Tanque.png)
<h4>Además se tiene que la concentración c es $c=\frac{m}{v}=\frac{\dot{m}}{\dot{v}}$    </h4>



 a) Establezca la la ecuación m vs t
 
 b) Halle la masa dentro del tanque en un tiempo de 11.3 minutos
 

<p style="text-align: justify;"> <span style="color:blue">Solución:</span></p>

a) La ecuación de balance material es: $$\frac{dm}{dt}=[Entrada]-[Salida]$$

$$\frac{dm}{dt}=\dot{m_1}-\dot{m_2}$$
$$\frac{dm}{dt}=c_1 v_1-c_2 v_2=45-c(t) v_2$$

Aquí vale lapena aclarar que $c_2=c(t)$ ya que el tanque está uniformemmente agotado y ,como $c(t)=\frac{m(t)}{25+2t}$

entonces nuestra ecuación diferencial se transforma en :

$$\frac{dm}{dt}=c_1 \dot{v_1}-c_2 \dot{v_2}=45-\frac{m(t)}{25+2t} \dot{v_2}$$


$$\frac{dm}{dt}=45-\frac{7}{25+2t} m(t)$$

Esta última ecuación es una ecuación diferencial LINEAL.

$$\frac{dm}{dt}+\frac{7}{25+2t} m=45$$

Para evaluar el factor integrante primero calculamos la integral $$\int \frac{7}{25+2t} dt=ln(25+2t)^{\frac{7}{2}}$$

Por lo tanto el factor integrante será:

$FI=e~^{\int ~\frac{7}{25+2t} dt}= e~^{ln(25+2t)^{\frac{7}{2}}}=(25+2t)^{\frac{7}{2}}$



Multiplicando la ecuación diferencial por el factor integrante se tiene:
$$\frac{d}{dt}\left ( m (25+2t \right )^{7/2})=45 (25+2t)^{7/2}$$


$$ m ~(25+2t ^{7/2})=45 \int (25+2t)^{7/2}dt$$

Haciendo la integral obtenemos la siguiente expresión:

$$ m~(25+2t  )^{7/2}=5 (25+2t)^{9/2}+k$$

De esta {ultima expresión despejamos la masa $m$

$$ m=5 (25+2t)+\frac {k}{(25+2t  )^{7/2}}$$


Ahora bien, con la condición inicial $m(t=0)=500~$ es posible evaluar la contante k:

$$500=5 \cdot 25+\frac{k}{25^{7/2}}$$

De donde se puede despejar la variable k:

$$k=(500-125)\cdot 25^{7/2}=29296875$$

La solución de la ecuación diferencial nos queda entonces:

$$ m=5 (25+2t)+\frac {29296875}{(25+2t  )^{7/2}}$$

b)EN el tiempo t=11.3 tendremos:

$$ m=5 (25+2\cdot 11.3)+\frac {29296875}{(25+2 \cdot 11.3  )^{7/2}}=277.37 ~~g$$
